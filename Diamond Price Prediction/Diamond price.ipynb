```python
def report(df,var):
    des = stats.describe(df[var])
    table_data=[
        ['Count',des.nobs],
        ['Min',des.minmax[0]],
        ['Max',des.minmax[1]],
        ['Mean',des.mean],
        ['Variance',des.variance],
        ['Skewness',des.skewness],
        ['Kurtosis',des.kurtosis]
    ]
    print(tabulate(table_data,headers=['Statistic','Value'],tablefmt='grid'))
```

```python
encode = ce.OrdinalEncoder(cols=['cut','color','clarity'],mapping=[
    {'col':'cut','mapping':{'Fair':1,'Good':2,'Very Good':3,'Premium':4,'Ideal':5}},
    {'col':'color','mapping':{'J':1,'I':2,'H':3,'G':4,'F':5,'E':6,'D':7}},
    {'col':'clarity','mapping':{'I1':1,'SI2':2,'SI1':3,'VS2':4,'VS1':5,'VVS2':6,'VVS1':7,'IF':8}}
])
```

```python
df_n = encode.fit_transform(df3)
df_n.head()
```

```python
import numpy as np
```

```python
df_n['log price'] = np.log(df_n['price'])
```

```python
df_n.head()
```

```python
df_n1 = df_n.drop(['price'],axis='columns')
df_n1.head()
```

# Model building and Performance Evaluating

```python
x = df_n1.drop('log price',axis='columns')
y = df_n1['log price']
```

```python
from sklearn.model_selection import train_test_split,cross_val_score,ShuffleSplit
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
len(y_test)
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mean_absolute_error(y_test,model.predict(x_test))
```

```python
mean_squared_error(y_test,model.predict(x_test))
```

```python
cv = ShuffleSplit(n_splits=5,test_size=0.2,random_state=42)
```

```python
cross_val_score(LinearRegression(),x,y,cv=cv)
```

```python
from scipy import stats

model_residuals = y_test - model.predict(x_test)
stats.probplot(model_residuals,dist='norm',plot=plt)
plt.show()
#plt.savefig('p-p plot',bbox_inches='tight')
```

```python
model_predict = model.predict(x_test)
model_residuals = y_test - model_predict

plt.scatter(model_predict,model_residuals)
plt.axhline(y=0,color='red',linestyle='--')
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Residuals vs. Predicted Values")
plt.show()
#plt.savefig('residuals vs fitted (linear regression)',bbox_inches='tight')
```

```python
def estimate_price(carat,cut,color,clarity,depth,table,x,y,z):
    X = np.zeros(len(x_train.columns))
    X[0] = carat
    X[1] = cut
    X[2] = color
    X[3] = clarity
    X[4] = depth
    X[5] = table
    X[6] = x
    X[7] = y
    X[8] = z

    log_price =  model.predict([X])[0]
    return np.exp(log_price)
```

```python
estimate_price(0.72,5,7,3,60.8,57,5.75,5.76,3.5)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRFRegressor
```

```python
models = [Ridge(),Lasso(),RandomForestRegressor(),SVR(),KNeighborsRegressor(),XGBRFRegressor()]
score = []

for model in models:
    scores = cross_val_score(model,x,y,cv=cv)
    scores = np.round(np.array(scores),3)

    score.append({
        'model':model,
        'scores':scores,
        'mean_score':np.mean(scores)
    })
```
